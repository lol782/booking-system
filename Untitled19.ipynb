```python
import json, os, requests
from difflib import get_close_matches
from fastapi import FastAPI, HTTPException
from fastapi.middleware.cors import CORSMiddleware
from pydantic import BaseModel
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.prompts import PromptTemplate
from google.colab import userdata
from pyngrok import ngrok
import nest_asyncio, uvicorn
import re

# ── Set up ──────────────────────────────────────────────────────────
os.environ["GOOGLE_API_KEY"] = userdata.get("GOOGLE_API_KEY")

app = FastAPI()

app.add_middleware(
    CORSMiddleware,
    allow_origins=["*"],
    allow_methods=["*"],
    allow_headers=["*"],
)

class Query(BaseModel):
    user_query: str
    token: str = None

# ── Enhanced LLM Prompt ──────────────────────────────────────────────────────
template = """
You are a friendly museum assistant. Help visitors with exhibits, timings, and bookings.

Important: When someone wants to book a ticket, ask for:
1. Which museum they want to visit
2. Their preferred date
3. Number of tickets (if not mentioned)

If they provide a museum name and show booking intent, respond with:
"Let me book [Museum Name] for you right away!"

Keep responses conversational and helpful.

User: {question}
Response:
"""
prompt = PromptTemplate(input_variables=["question"], template=template)

llm = ChatGoogleGenerativeAI(model="models/gemini-2.0-flash", temperature=0.4)

# ── Booking Logic ──────────────────────────────────────────────────
BACKEND_URL = "https://2b3ed9c85fcd.ngrok-free.app/lol"  # Update to your deployed Django backend URL

def get_museum_id_by_name(name_query, token):
    """Get museum ID by matching name from Django backend."""
    try:
        response = requests.get(
            f"{BACKEND_URL}/api/browse/",

        )

        if response.status_code != 200:
            print(f"API Error: {response.status_code} - {response.text}")
            return None, None

        museum_list = response.json()
        names = [museum["name"] for museum in museum_list]

        # Match closest name
        best_match = get_close_matches(name_query, names, n=1, cutoff=0.5)
        if not best_match:
            return None, None

        for museum in museum_list:
            if museum["name"] == best_match[0]:
                return museum["museum_id"], museum["name"]
        return None, None
    except Exception as e:
        print(f"Error fetching museums: {e}")
        return None, None

def extract_museum_name(user_input):
    """Try to extract museum name from user input."""
    # Common patterns for museum booking requests
    patterns = [
        r"book\s+(.+?)(?:\s+for|\s+on|\s+ticket|$)",
        r"reserve\s+(.+?)(?:\s+for|\s+on|\s+ticket|$)",
        r"visit\s+(.+?)(?:\s+museum|\s+on|\s+for|$)",
        r"ticket\s+for\s+(.+?)(?:\s+museum|\s+on|\s+for|$)",
        r"(.+?)\s+museum",
    ]

    for pattern in patterns:
        match = re.search(pattern, user_input.lower())
        if match:
            return match.group(1).strip()
    return user_input  # Return full input if no pattern matches

def get_available_museums(token):
    """Get list of available museums for fallback booking."""
    try:
        response = requests.get(
            f"{BACKEND_URL}/api/browse/"
        )
        if response.status_code == 200:
            return response.json()
        return []
    except:
        return []

def book_ticket(token, user_input):
    """Book a ticket through Django backend."""
    try:
        headers = {"Authorization": f"Bearer {token}"}

        # Extract potential museum name from user input
        museum_name_query = extract_museum_name(user_input)

        museum_id, matched_name = get_museum_id_by_name(museum_name_query, token)

        # If no specific museum found, try to book the first available museum
        if not museum_id:
            museums = get_available_museums(token)
            if museums:
                museum_id = museums[0]["museum_id"]
                matched_name = museums[0]["name"]
                print(f"No specific museum found, booking first available: {matched_name}")

        if museum_id:
            book_url = f"{BACKEND_URL}/api/book_museum/{museum_id}/"
            book_res = requests.post(book_url, headers=headers)

            if book_res.status_code in [200, 201]:
                response_data = book_res.json()
                return f"🎉 Perfect! Your museum ticket has been booked successfully!\n\n📋 **Booking Confirmation:**\n- Museum: **{matched_name}**\n- Booking ID: #{response_data.get('booking_id')}\n- Visit Date: {response_data.get('visit_date')}\n- Ticket Type: {response_data.get('ticket_type', 'Standard')}\n\n🎫 Please save this booking ID for your records. Have a wonderful visit!"
            else:
                try:
                    error_data = book_res.json()
                    error_msg = error_data.get('error', 'Unknown error')
                except:
                    error_msg = book_res.text
                return f"❌ Sorry, I couldn't complete your booking: {error_msg}\n\nPlease try again or contact support if the issue persists."
        else:
            return f"❌ I'm sorry, I couldn't find any available museums to book. Please make sure museums are available in the system."
    except Exception as e:
        return f"❌ Booking failed due to a technical issue: {str(e)}\n\nPlease try again in a moment."

def should_attempt_booking(user_query, response):
    """Determine if we should attempt to book based on user query and LLM response."""
    booking_keywords = ["book", "ticket", "reserve", "booking", "visit"]

    query_lower = user_query.lower()
    response_lower = response.lower()

    # Check if user explicitly wants to book
    user_wants_booking = any(keyword in query_lower for keyword in booking_keywords)

    # Check if LLM response indicates booking processing
    llm_processing_booking = any(phrase in response_lower for phrase in [
        "let me book", "let me process", "booking request", "process your booking"
    ])

    # Attempt booking if either condition is true
    return user_wants_booking or llm_processing_booking

# ── Chat Endpoint ──────────────────────────────────────────────────
@app.post("/chat")
async def chat_bot(q: Query):
    try:
        print("User Query:", q.user_query)
        print("Token provided:", bool(q.token))

        # Get LLM response
        response = llm.invoke(prompt.format(question=q.user_query)).content
        print("LLM Response:", response)

        # Check if we should attempt booking
        if should_attempt_booking(q.user_query, response):
            if q.token:
                print("Attempting to book...")
                booking_msg = book_ticket(q.token, q.user_query)
                return {"answer": booking_msg}  # Return only booking result
            else:
                return {"answer": "🔐 **To book your museum ticket, please log in first.**\n\nOnce you're logged in, just let me know which museum you'd like to visit and I'll handle the booking for you!"}

        return {"answer": response}

    except Exception as e:
        print("Error:", str(e))
        raise HTTPException(status_code=500, detail=str(e))

# ── Additional helper endpoint ──────────────────────────────────────
@app.get("/museums")
async def list_museums():
    """Endpoint to list available museums (for testing)."""
    try:
        response = requests.get(f"{BACKEND_URL}/api/browse/")
        if response.status_code == 200:
            return response.json()
        return {"error": "Could not fetch museums"}
    except Exception as e:
        return {"error": str(e)}
```
